```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np

# Function to minimize (quadratic function)
def quadratic_function(x):
    return x**2 + 5 * x + 6
def polynomial_function(x):
    return x**4 + x**3 - 5*x**2

# Gradient descent optimization function
def gradient_descent(lr, num_epochs=100, initial_position=-10):
    x = torch.tensor(initial_position, requires_grad=True, dtype=torch.float)
    optimizer = optim.SGD([x], lr=lr)

    positions = []

    for epoch in range(num_epochs):
        y = quadratic_function(x)
        optimizer.zero_grad()
        y.backward()
        optimizer.step()

        # Save the current position for plotting
        positions.append(x.item())

        # Print the position every 10 epochs
        if (epoch + 1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], Position: {x.item():.4f}, Function Value: {y.item():.4f}')

    # Plot the optimization trajectory
    x_values = np.linspace(-15, 5, 100)
    y_values = quadratic_function(x_values)
    plt.plot(x_values, y_values, label='Quadratic Function')
    plt.scatter(positions, [quadratic_function(pos) for pos in positions], color='red', marker='o', label='Optimization Trajectory')
    plt.xlabel('Position (x)')
    plt.ylabel('Function Value')
    plt.title('Gradient Descent Optimization')
    plt.legend()
    plt.show()

# Example usage:
# lr = 0.1
# gradient_descent(lr, num_epochs=100)
```

```python
lr = 0.01
gradient_descent(lr)  
```
